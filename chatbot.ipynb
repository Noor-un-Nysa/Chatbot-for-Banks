```python
from sklearn.preprocessing import LabelEncoder

# Get intents as labels
labels = training_df['intent'].tolist()

# Convert string labels to numeric
le = LabelEncoder()
y = le.fit_transform(labels)  # e.g., 'greeting' → 0, 'personal_balance' → 1, etc.

print("Example encoded labels:", y[:])
```

```python
from sklearn.neural_network import MLPClassifier

# Generate embeddings for all training texts
training_texts = training_df['text'].tolist()
embeddings = model.encode(training_texts)

# Train the classifier
clf = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
clf.fit(embeddings, y)
```

```python
def generate_response(intent, user_input, user_name="Ali"):
    if intent == "greeting":
        return "Hello! How can I help you today?"

    elif intent == "goodbye":
        return "Goodbye! Have a nice day."

    elif intent == "personal_balance":
        # Lookup user balance
        balance = users_df.loc[users_df['name']==user_name, 'balance'].values[0]
        return f"Your account balance is: {balance} PKR."

    elif intent == "personal_card_info":
        card = users_df.loc[users_df['name']==user_name, 'card_number'].values[0]
        expiry = users_df.loc[users_df['name']==user_name, 'expiry'].values[0]
        return f"Your card number is {card} and expires on {expiry}."

    elif intent == "personal_complaint":
        status = users_df.loc[users_df['name']==user_name, 'complaint_status'].values[0]
        return f"Your complaint status is: {status}."

    elif intent == "nearest_branch":
        # Ask for location in input (simplest way)
        location = input("Bot: Where are you?\nYou: ")
        neighbors = ", ".join(branches.get(location, []))
        manager = managers.get(location, "Manager info not available")
        return f"Nearest branches: {neighbors}\nBranch manager: {manager}"

    elif intent == "general_faq":
        # Simple example: static FAQ dictionary
        faq_dict = {
            "working hours": "Our working hours are 9 AM to 5 PM, Monday to Friday.",
            "loan": "You can apply for a loan online or at any branch."
        }
        # naive matching
        for key, answer in faq_dict.items():
            if key in user_input.lower():
                return answer
        return "Sorry, I don't know the answer to that question."

    else:
        return "Sorry, I didn't understand. Can you rephrase?"
```

```python
# Dictionary to temporarily store user's name for session
session = {"user_name": None}

def generate_response(intent, user_input):
    # For intents that need a name
    if intent in ["personal_balance", "personal_card_info", "personal_complaint"]:
        # Ask for name if we don't have it yet
        if not session["user_name"]:
            session["user_name"] = input("Bot: Sure! May I have your name?\nYou: ")

        # Check if name exists in users_df
        if session["user_name"] not in users_df['name'].values:
            session["user_name"] = None  # reset
            return "Sorry, we don’t have a customer with that name."

        # Name exists → handle intent
        user_name = session["user_name"]
        if intent == "personal_balance":
            balance = users_df.loc[users_df['name']==user_name, 'balance'].values[0]
            return f"{user_name}, your account balance is: {balance} PKR."

        elif intent == "personal_card_info":
            card = users_df.loc[users_df['name']==user_name, 'card_number'].values[0]
            expiry = users_df.loc[users_df['name']==user_name, 'expiry'].values[0]
            return f"{user_name}, your card number is {card} and expires on {expiry}."

        elif intent == "personal_complaint":
            status = users_df.loc[users_df['name']==user_name, 'complaint_status'].values[0]
            return f"{user_name}, your complaint status is: {status}."

    # For intents that don’t need a name
    elif intent == "greeting":
        return "Hello! How can I help you today?"

    elif intent == "goodbye":
        return "Goodbye! Have a nice day."

    elif intent == "nearest_branch":
        location = input("Bot: Where are you?\nYou: ")
        neighbors = ", ".join(branches.get(location, []))
        manager = managers.get(location, "Manager info not available")
        return f"Nearest branches: {neighbors}\nBranch manager: {manager}"

    elif intent == "general_faq":
        faq_dict = {
            "working hours": "Our working hours are 9 AM to 5 PM, Monday to Friday.",
            "loan": "You can apply for a loan online or at any branch.",
            "online banking": "Yes, we provide online banking services.",
            "open account": "You can open a new account at any branch or online."
        }
        for key, answer in faq_dict.items():
            if key in user_input.lower():
                return answer
        return "Sorry, I don't know the answer to that question."

    else:
        return "Sorry, I didn't understand. Can you rephrase?"
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() in ["bye", "exit", "quit"]:
        print("Bot: Goodbye!")
        break

    user_emb = model.encode([user_input])
    pred_label = clf.predict(user_emb)
    intent = le.inverse_transform(pred_label)[0]

    response = generate_response(intent, user_input)
    print("Bot:", response)
```
